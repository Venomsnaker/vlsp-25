```python
from utility import load_img_paths
import os

output_path = '../data/output/finetuned'
os.makedirs(output_path, exist_ok=True)

yolo_path = '../data/yolo/yolov8n_finetuned.pt'
img_paths = load_img_paths('../data/train_data/train_images')
```

```python
from ultralytics import YOLO
import cv2
import os

model = YOLO(yolo_path)

for img_path in img_paths:
    img_output_path = os.path.join(output_path, 'modified_' + os.path.basename(img_path))
    img = cv2.imread(img_path)
    
    results = model(img_path)
    for r in results:
        boxes = r.boxes.xyxy.cpu().numpy()  # [x1, y1, x2, y2]
        class_ids = r.boxes.cls.cpu().numpy().astype(int)
        confidences = r.boxes.conf.cpu().numpy()
        
        for box, cls_id, conf in zip(boxes, class_ids, confidences):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[cls_id]} {conf:.2f}"

            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Label
            # (w, h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
            # cv2.rectangle(img, (x1, y1 - h - 4), (x1 + w, y1), (0, 255, 0), -1)

            cv2.putText(img, label, (x1, y1 - 4), cv2.FONT_HERSHEY_SIMPLEX,
                        0.6, (0, 0, 0), 1)
        
    cv2.imwrite(img_output_path, img)
```
